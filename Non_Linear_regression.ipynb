```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(-5.0,5.0,0.1)
```

```python
y=2*(x)+3
y_noise=2*np.random.normal(size=x.size)
ydata=y+y_noise
```

```python
plt.plot(x,ydata,'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show
```

```python
y=np.power(x,2)
y_noise=2*np.random.normal(size=x.size)
ydata=y+y_noise
plt.plot(x,ydata,'bo')
plt.plot(x,y,'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
